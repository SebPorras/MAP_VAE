```python
import evoVAE.utils.seq_tools as st
import pandas as pd
```

This notebook will use travis to simulate sequences under and evolutionary model 

```python
out_dir = "/Users/sebs_mac/uni_OneDrive/honours/data/gb1/travis_simulations/"
tree_dir = "/Users/sebs_mac/uni_OneDrive/honours/data/gb1/trees/"
ancestor_dir = "/Users/sebs_mac/uni_OneDrive/honours/data/gb1/ancestors/seqs/"
```

```python
commands = []
for i in range(1, 31):

    aln = st.read_aln_file(f"{ancestor_dir}anc_tree_{i}_ancestors.fa")
    n0 = aln[aln["id"] == f"N0_gb1_tree_{i}"]["sequence"][0]
    tree = f"{tree_dir}run_{i}.treefile"
    outfile = f"{out_dir}run_{i}_sim.fasta"
    c = f"java -Xmx16g -jar {out_dir}travis.jar {n0} -nwk {tree} -out {outfile}"
    commands.append(c)

    
```

```python
commands
```

```python
import subprocess

for command in commands:
    # Execute the shell command
    process = subprocess.Popen(command, shell=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE)
    
    # Get the output and error (if any)
    stdout, stderr = process.communicate()
    
    # Print the output and error
    if stdout:
        print("Output:\n", stdout.decode())
    if stderr:
        print("Error:\n", stderr.decode())
```

```python
alns = []
for i in range(1,31):
    aln = st.read_aln_file(f"{out_dir}run_{i}_sim.fasta")
    aln = aln[~aln["id"].str.contains("/")]
    alns.append(aln)


all = pd.concat(alns)
```

```python
all.drop_duplicates(subset=["sequence"], inplace=True)
```
